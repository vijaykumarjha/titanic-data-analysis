```python
titanic_data.drop("Parents/Children Aboard" , axis = 1,inplace = True)
```

```python
titanic_data.head(5)
```

```python
sex = pd.get_dummies(titanic_data["Sex"],drop_first = True)
sex.head()
```

```python
pclass= pd.get_dummies(titanic_data["Pclass"],drop_first=True)
pclass.head()
```

```python
ssa= pd.get_dummies(titanic_data["Siblings/Spouses Aboard"],drop_first=True)
ssa.head()
```

```python
titanic_data=pd.concat([titanic_data,sex,pclass,ssa],axis =1)
titanic_data.head()
```

```python
titanic_data.drop(["Pclass","Name","Sex","Age","Siblings/Spouses Aboard","Fare"],axis = 1,inplace = True)
```

```python
titanic_data.head()
```

##Train data

```python
x=titanic_data.drop("Survived",axis = 1)
y = titanic_data["Survived"]
```

```python

from sklearn.model_selection import train_test_split
```

```python
 x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.33, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
predictions =logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import sklearn
import seaborn as sns
import math
```

```python
dataset = pd.read_csv(r"C:\Users\VIJAY KR JHA\Desktop\python\Social_Network_Ads.csv")
dataset.head()
```

```python
x = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:,4].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
 x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score 
```

```python
accuracy_score(y_test,y_pred)
```
